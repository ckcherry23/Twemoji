```python
import os
train_fnames = os.listdir("/content/gdrive/My Drive/CS3244/Train/")
val_fnames = os.listdir("/content/gdrive/My Drive/CS3244/Validation/")
test_fnames = os.listdir("/content/gdrive/My Drive/CS3244/Test/")
```

```python
#Read dataframe for only 1 (work on merging all pandas dataframe TBD)
df = pd.concat(map(lambda x: pd.read_csv(f"/content/gdrive/My Drive/CS3244/Train/{x}", header=None), train_fnames), ignore_index=True)
#Drop first column and set col names
df = df.iloc[: , 1:]
df.columns = ["id", "annotations", "text", "hashtag"]
df = df.dropna(subset=['text']).fillna("")
df
```

```python
##Number of annotations in train. Should be 20
print(df['annotations'].nunique())
```

```python
#Get description of annotations, data looks relatively balanced
print(df["annotations"].value_counts())
#Get number of rows with hashtag
print(len(df[df["hashtag"] != ""]))
```

```python
#Removing emoji and hashtag from text
!pip install emoji
import emoji
import re

df_new = df.copy()
def strip_emoji(text):
    new_text =  re.sub(":([^,]*):","",emoji.demojize(text))
    return new_text
  
df_new["text"] = df_new["text"].apply(lambda x: strip_emoji(x))
df_new
```

```python
#Update hashtag column
from ast import literal_eval
def extract_hashtag(text):
  if text != "":
    lst = [i['tag'] for i in literal_eval(text)]
    return " ".join(lst)
  return text

df_new["hashtag"] = df_new["hashtag"].apply(lambda x: extract_hashtag(x))
#Sanity check
df_new[df_new["hashtag"] != ""]
```

```python
#Remove hashtag and links from hashtag from existing text
  
hashtag_df = df_new[df_new["hashtag"] != ""].copy()
hashtag_df["text"] = hashtag_df["text"].apply(lambda x: re.sub(' #([^,]*)', "", x))

#Merge two dataframe
df_v2 = pd.concat([hashtag_df, df_new[df_new["hashtag"] == ""]]).reset_index().drop(columns=['id', 'index'])
df_v2["text"] = df_v2["text"].apply(lambda x: re.sub('//t.co/([^,]*)', "", x))
df_v2
```

```python
# Download nltk resources
import nltk
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download('omw-1.4')
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()
lem.lemmatize("freezing")
```

```python
#Tokenization of the text, use Tweet Tokenizer to remove handles of tweets 
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer 
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import TweetTokenizer
import string

tknzr = TweetTokenizer(strip_handles=True, reduce_len=True)
stpwrds = stopwords.words("english")
stpwrds.append("rt")
lem = WordNetLemmatizer()

def preprocess(text):
  remove_tags = " ".join([i for i in text.split() if i[0] != "@"])
  removed_punctuation = "".join([i for i in remove_tags if i not in string.punctuation])
  lower_case = removed_punctuation.lower()
  tokens = tknzr.tokenize(lower_case)
  removed_stopwords = [i for i in tokens if i not in stpwrds]
  lemm_text = [lem.lemmatize(word) for word in removed_stopwords]
  return lemm_text

df_v2["tokens"] = df_v2["text"].apply(lambda x: preprocess(x))
df_v2
```

```python
df_v2["len"] = df_v2["tokens"].apply(lambda x: len(x))
df_v3 = df_v2[df_v2["len"] != 0]
print(20000 - df_v3["annotations"].value_counts())
```
